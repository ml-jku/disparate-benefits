```python
import os
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from matplotlib.lines import Line2D
import seaborn as sns
sns.set_theme()

from source.constants import PLOTS_PATH

os.makedirs(PLOTS_PATH, exist_ok=True)
```

```python
# data taken from analyze_face_detection_postpro.iypnb
val_fair = [0.169, 0.149, 0.111]
accs_spd_m = [0.794, 0.816, 0.818, 0.818, 0.789]
accs_spd_s = [0.003, 0.002, 0.001, 0.001, 0.005]
spds_m = [0.173, 0.194, 0.176, 0.057, 0.056]
spds_s = [0.007, 0.004, 0.011, 0.003, 0.024]
accs_eod_m = [0.794, 0.816, 0.818, 0.815, 0.792]
accs_eod_s = [0.003, 0.002, 0.001, 0.002, 0.005]
eods_m = [0.153, 0.171, 0.157, 0.067, 0.055]
eods_s = [0.012, 0.004, 0.014, 0.006, 0.021]
accs_aod_m = [0.794, 0.816, 0.818, 0.816, 0.793]
accs_aod_s = [0.003, 0.002, 0.001, 0.002, 0.005]
aods_m = [0.113, 0.129, 0.114, 0.062, 0.054]
aods_s = [0.008, 0.004, 0.012, 0.002, 0.015]
```

```python
c_member = "gray"
c_ens = "C3"
c_arrow = "black"

fig, axs = plt.subplots(1, 3, figsize=(12, 3))

axs[0].errorbar(accs_spd_m[0], spds_m[0], xerr=accs_spd_s[0], yerr=spds_s[0], fmt='o', label='Members', color=c_member, zorder=50)
axs[0].errorbar(accs_spd_m[1], spds_m[1], xerr=accs_spd_s[1], yerr=spds_s[1], fmt='o', label='Deep Ensemble', color=c_ens, zorder=50)
axs[0].errorbar(accs_spd_m[2], spds_m[2], xerr=accs_spd_s[2], yerr=spds_s[2], fmt='o', color=c_ens, zorder=50)
axs[0].errorbar(accs_spd_m[3], spds_m[3], xerr=accs_spd_s[3], yerr=spds_s[3], fmt='o', color=c_ens, zorder=50)
axs[0].errorbar(accs_spd_m[4], spds_m[4], xerr=accs_spd_s[4], yerr=spds_s[4], fmt='o', color=c_member, zorder=50)
axs[0].set_ylabel('SPD ($\\leftarrow$)')
axs[0].annotate('', xy=(accs_spd_m[2], spds_m[2]), xytext=(accs_spd_m[1], spds_m[1]),
                 arrowprops=dict(arrowstyle='-|>', color=c_arrow, linestyle='solid'), zorder=99, alpha=0.5)
axs[0].annotate('', xy=(accs_spd_m[3], spds_m[3]), xytext=(accs_spd_m[1], spds_m[1]),
                 arrowprops=dict(arrowstyle='-|>', color=c_arrow, linestyle='solid'), zorder=99, alpha=0.5)
axs[0].annotate('', xy=(accs_spd_m[4], spds_m[4]), xytext=(accs_spd_m[0], spds_m[0]),
                 arrowprops=dict(arrowstyle='-|>', color=c_arrow, linestyle='solid'), zorder=99, alpha=0.5)
xmin, xmax = axs[0].get_xlim()
axs[0].hlines(0.05, xmin, xmax, color='black', linestyle='dashed', alpha=0.25)
axs[0].hlines(val_fair[0], xmin, xmax, color='black', linestyle='dotted', alpha=0.25)
axs[0].set_xlim(xmin, xmax)

bar1 = axs[1].errorbar(accs_eod_m[0], eods_m[0], xerr=accs_eod_s[0], yerr=eods_s[0], fmt='o', label='Members', color=c_member, zorder=50)
bar2 = axs[1].errorbar(accs_eod_m[1], eods_m[1], xerr=accs_eod_s[1], yerr=eods_s[1], fmt='o', label='Deep Ensemble', color=c_ens, zorder=50)
axs[1].errorbar(accs_eod_m[2], eods_m[2], xerr=accs_eod_s[2], yerr=eods_s[2], fmt='o', color=c_ens, zorder=50)
axs[1].errorbar(accs_eod_m[3], eods_m[3], xerr=accs_eod_s[3], yerr=eods_s[3], fmt='o', color=c_ens, zorder=50)
axs[1].errorbar(accs_eod_m[4], eods_m[4], xerr=accs_eod_s[4], yerr=eods_s[4], fmt='o', color=c_member, zorder=50)
axs[1].set_ylabel('EOD ($\\leftarrow$)')
axs[1].annotate('', xy=(accs_eod_m[2], eods_m[2]), xytext=(accs_eod_m[1], eods_m[1]),
                 arrowprops=dict(arrowstyle='-|>', color=c_arrow, linestyle='solid'), zorder=99)
axs[1].annotate('', xy=(accs_eod_m[3], eods_m[3]), xytext=(accs_eod_m[1], eods_m[1]),
                 arrowprops=dict(arrowstyle='-|>', color=c_arrow, linestyle='solid'), zorder=99)
axs[1].annotate('', xy=(accs_eod_m[4], eods_m[4]), xytext=(accs_eod_m[0], eods_m[0]),
                 arrowprops=dict(arrowstyle='-|>', color=c_arrow, linestyle='solid'), zorder=99)
xmin, xmax = axs[1].get_xlim()
axs[1].hlines(0.05, xmin, xmax, color='black', linestyle='dashed', alpha=0.25)
axs[1].hlines(val_fair[1], xmin, xmax, color='black', linestyle='dotted', alpha=0.25)
axs[1].set_xlim(xmin, xmax)

axs[2].errorbar(accs_aod_m[0], aods_m[0], xerr=accs_aod_s[0], yerr=aods_s[0], fmt='o', label='Members', color=c_member, zorder=50)
axs[2].errorbar(accs_aod_m[1], aods_m[1], xerr=accs_aod_s[1], yerr=aods_s[1], fmt='o', label='Deep Ensemble', color=c_ens, zorder=50)
axs[2].errorbar(accs_aod_m[2], aods_m[2], xerr=accs_aod_s[2], yerr=aods_s[2], fmt='o', color=c_ens, zorder=50)
axs[2].errorbar(accs_aod_m[3], aods_m[3], xerr=accs_aod_s[3], yerr=aods_s[3], fmt='o', color=c_ens, zorder=50)
axs[2].errorbar(accs_aod_m[4], aods_m[4], xerr=accs_aod_s[4], yerr=aods_s[4], fmt='o', color=c_member, zorder=50)
axs[2].set_ylabel('AOD ($\\leftarrow$)')
axs[2].annotate('', xy=(accs_aod_m[2], aods_m[2]), xytext=(accs_aod_m[1], aods_m[1]),
                 arrowprops=dict(arrowstyle='-|>', color=c_arrow, linestyle='solid'), zorder=99)
axs[2].annotate('', xy=(accs_aod_m[3], aods_m[3]), xytext=(accs_aod_m[1], aods_m[1]),
                 arrowprops=dict(arrowstyle='-|>', color=c_arrow, linestyle='solid'), zorder=99)
axs[2].annotate('', xy=(accs_aod_m[4], aods_m[4]), xytext=(accs_aod_m[0], aods_m[0]),
                 arrowprops=dict(arrowstyle='-|>', color=c_arrow, linestyle='solid'), zorder=99)
xmin, xmax = axs[2].get_xlim()
axs[2].hlines(0.05, xmin, xmax, color='black', linestyle='dashed', alpha=0.25)
axs[2].hlines(val_fair[2], xmin, xmax, color='black', linestyle='dotted', alpha=0.25)
axs[2].set_xlim(xmin, xmax)

arrow_handle = Line2D([0], [0], marker=r'$\rightarrow$', color=c_arrow, markersize=10, linestyle='None')
axs[2].legend(handles=[bar1, bar2, arrow_handle], labels=['Members', 'Deep Ensemble', 'Post-processing'], fontsize="small", loc="center", ncol=1)

for ax in axs:
    ax.set_xlabel('Accuracy ($\\rightarrow$)')
    ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{x:.3f}"))

plt.tight_layout()
plt.savefig(os.path.join(PLOTS_PATH, f"postpro.pdf"))
plt.show()
```
